```python
#Import libries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
plt.style.use('fivethirtyeight')
```

```python
FCA = pd.read_csv('FCA.MI.csv')
```

```python
FCA
```

```python
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = FCA['Adj Close'].rolling(window=30).mean()
SMA30
```

```python
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = FCA['Adj Close'].rolling(window=100).mean()
SMA100
```

```python
plt.figure(figsize=(12.5,4.5))
plt.plot(FCA['Adj Close'], label='FCA')
plt.plot(SMA30['Adj Close'], label='FCA SMA30')
plt.plot(SMA100['Adj Close'], label='FCA SMA100')
plt.title('Adj Price')
plt.xlabel('One year windows')
plt.ylabel('Price in euro')
plt.legend(loc='upper left')
plt.show()

```

```python
data = pd.DataFrame()
data['Date'] = FCA['Date']
data['FCA'] = FCA['Adj Close']
data['SMA30'] = SMA30['Adj Close']
data['SMA100'] = SMA100['Adj Close']
data.set_index('Date', inplace=True)
data
```

```python
def buy_sell(data):
    sigPriceBuy = []
    sigPriceSell = []
    flag = -1
    
    
    for i in range(len(data)):
        if data['SMA30'][i] > data['SMA100'][i]:
            if flag != 1:
                sigPriceBuy.append(data['FCA'][i])
                sigPriceSell.append(np.nan)
                flag = 1
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)

        elif data['SMA30'][i] < data['SMA100'][i]:
            if flag != 0:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(data['FCA'][i])
                flag = 0
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        else:
            sigPriceBuy.append(np.nan)
            sigPriceSell.append(np.nan)  
    
    return (sigPriceBuy, sigPriceSell)
```

```python
buy_sell = buy_sell(data)
data['Buy_Signal_Price'] = buy_sell[0]
data['Sell_Signal_Price'] = buy_sell[1]
```

```python
data
```

```python
fig = plt.figure(figsize=(40.6,20.6))
plt.plot(data['FCA'], label='FCA', alpha=0.35)
plt.plot(data['SMA30'], label='FCA SMA30', alpha=0.35)
plt.plot(data['SMA100'], label='FCA SMA100', alpha=0.35)
plt.scatter(data.index, data['Buy_Signal_Price'], label = 'Buy', marker='^', color= 'green')
plt.scatter(data.index, data['Sell_Signal_Price'], label = 'Sell', marker='v', color= 'red')
plt.title('Adj FCA Price with Buy and Sell Signals')
plt.xlabel('From 25 May 2019 - to Now')
plt.ylabel('Price in euro')
plt.xticks(rotation=30, pad=20)
plt.legend(loc='upper left')
plt.show()

```
